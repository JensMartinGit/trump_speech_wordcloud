# Trump 6 January Speech Wordcloud

I saw a wordcloud of US president Joe Biden's inauguration speech today and simply couldn't resist the temptation to create another one of Donald Trump's speech from 6 January to highlight the contrast. ;)

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
from PIL import Image
```

```python
# Get text corpus from file
with open('trump_speech_6_jan.txt', 'r', encoding='utf-8') as textfile:
    corpus = textfile.read()
```

```python
# Create a very basic wordcloud

# Generate word cloud image
wordcloud = WordCloud(height=400, min_word_length=3).generate(corpus)

# Display wordcloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off');
```

```python
# Create wordcloud with Trump picture mask

# Get image for mask from file
mask = np.array(Image.open('Trump-space-invaders-derivative2.png'))

# Set stopwords
stopwords = set(STOPWORDS)
stopwords.add('don')

# Generate wordcloud with mask
word_cloud2 = WordCloud(width=700, height=700, min_word_length=3, min_font_size=5, 
                        max_font_size=50, mask=mask, background_color='white', 
                        stopwords=stopwords, colormap='gist_heat'
                       ).generate(corpus)

# Display wordcloud with overlapping image
plt.figure(figsize=(10,10))
plt.imshow(word_cloud2, interpolation='bilinear')
plt.imshow(mask, alpha=0.6)
plt.axis('off');
```
