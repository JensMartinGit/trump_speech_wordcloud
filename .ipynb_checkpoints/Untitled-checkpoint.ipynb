# Trump 6 January Speech Wordcloud

I saw a wordcloud of Joe Biden's inauguration speech and simply couldn't resist the temptation to create another one of Donald Trump's speech from 6 January to highlight the contrast. ;)

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud
from PIL import Image
```

```python
# Get text corpus from file
with open('trump_speech_6_jan.txt', 'r', encoding='utf-8') as textfile:
    corpus = textfile.read()
```

```python
# Create a very basic wordcloud

# Generate word cloud image
wordcloud = WordCloud(height=400, min_word_length=3).generate(corpus)

# Display wordcloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off');
```
